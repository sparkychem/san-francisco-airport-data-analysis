```python
import pandas as pd
import numpy as np
from geopy import distance
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('air travel dataset.xlsx - Air travel data.csv')
df.head()
```

```python
df.info()
```

```python
df[['From', 'To']] = df.Route.apply(lambda x: pd.Series(str(x).split('-')))
df.head()
```

```python
del df['Route']
df.head()
```

```python
def distances(latitude, longitude):
    sfo = (37.61799, -122.370943)
    point = (longitude, latitude)
    
    return distance.distance(sfo, point).km
```

```python
print((distances(-99.080804, 19.431302)), 'Km')
```

```python
df['Distance (Km)'] = df.apply(lambda df: distances(df['Geometry Coordinates 1 0'], df['Geometry Coordinates 1 1'])
                          , axis=1)
```
